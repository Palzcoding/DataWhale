### 2.3 特征观察与处理
我们对特征进行一下观察，可以把特征大概分为两大类：  
数值型特征：Survived ，Pclass， Age ，SibSp， Parch， Fare，其中Survived， Pclass为离散型数值特征，Age，SibSp， Parch， Fare为连续型数值特征  
文本型特征：Name， Sex， Cabin，Embarked， Ticket，其中Sex， Cabin， Embarked， Ticket为类别型文本特征，数值型特征一般可以直接用于模型的训练，但有时候为了模型的稳定性及鲁棒性会对连续变量进行离散化。文本型特征往往需要转换成数值型特征才能用于建模分析。

```python
train.head(4)
```

数值型特征 (Numerical Features)：

Survived：表示乘客是否幸存的特征。通常是二进制特征（0表示未幸存，1表示幸存）。
Pclass：表示船舱等级的特征，通常是1, 2, 3等离散值。
Age：表示乘客年龄的特征，是连续的数值型特征。
SibSp：表示乘客在船上的兄弟姐妹/配偶数量的特征。
Parch：表示乘客在船上的父母/子女数量的特征。
Fare：表示乘客票价的特征，是连续的数值型特征。
离散型数值特征 (Discrete Numerical Features)：

离散型数值特征。这些特征取有限个不同的值，通常是整数。例如，Pclass的值只能是1, 2, 或3。

连续型数值特征 (Continuous Numerical Features)
这些特征可以取任意数值范围内的值，通常是浮点数。
文本型特征 

(Text Features)：
Sex，Cabin，Embarked，和Ticket是类别型文本特征。这些特征取有限个不同的文本值，表示不同的类别。

模型的稳定性及鲁棒性 (Model Stability and Robustness)：
模型的稳定性：指模型在面对不同数据集（例如训练集和测试集）时，其表现（例如预测准确率）的一致性。一个稳定的模型在不同的数据集上表现相近。
模型的鲁棒性：指模型在面对数据中的噪音或异常值时，仍能保持良好性能的能力。一个鲁棒的模型不容易被少量异常数据影响。

离散化 (Discretization)：
将连续型数值特征转换为离散型数值特征。比如，将年龄按范围划分为几个年龄段（0-10岁，11-20岁，...）以便于某些机器学习算法更好地处理。
数值型特征转换 (Conversion of Numerical Features)：

通常，数值型特征可以直接用于机器学习模型的训练。然而，为了提高模型的稳定性和鲁棒性，有时候需要对连续型数值特征进行离散化。

文本型特征转换 (Conversion of Text Features)：
文本型特征需要转换成数值型特征才能用于机器学习模型。例如，将类别型文本特征用One-Hot编码（one-hot encoding）转换为数值特征。

```python
## 用sklearn.preprocessing.LabelEncoder() 将文本值转换为数值标签
import pandas as pd
train = pd.read_csv('train.csv')
from sklearn import preprocessing
for feat in ['Cabin','Embarked']:
    map_dict = dict(zip(train[feat].unique(),range(train[feat].nunique())))  # unique 和nunique没有s，而且后面加（）表示function
    train[feat+'_label_encode'] = train[feat].map(map_dict)
    
    label = preprocessing.LabelEncoder()
    train[feat+'_label_encode'] = label.fit_transform(train[feat].astype(str)) # astype(str)转化成str，将所有数值都转换成str避免出错
train.head(3)
    
```

```python
## One-hot 编码 （get_dummies）
pd.get_dummies(train,
               columns=['Sex','Cabin','Embarked'])
```
