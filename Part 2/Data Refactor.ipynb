### 2.4 数据的合并

#### 2.4.1 任务一：将data文件夹里面的所有数据都载入，观察数据的之间的关系

```python
#写入代码
df2 = pd.read_csv('train-left-down.csv')
df3 = pd.read_csv('train-right-up.csv')
df4 = pd.read_csv('train-right-down.csv')
```

```python
#写入代码
df1.head(3)
df2.head(3)
df3.head(3)
df4.head(3)
```

【提示】结合之前我们加载的train.csv数据，大致预测一下上面的数据是什么

#### 2.4.2：任务二：使用concat方法：将数据train-left-up.csv和train-right-up.csv横向合并为一张表，并保存这张表为result_up

```python
df_up = pd.concat([df1,df3],axis=1)
df_up
```

#### 2.4.3 任务三：使用concat方法：将train-left-down和train-right-down横向合并为一张表，并保存这张表为result_down。然后将上边的result_up和result_down纵向合并为result。

```python
list_down = [df2,df4]
df_down = pd.concat(list_down,axis=1)
df_down
```

```python
list_result = [df_up, df_down]
result = pd.concat(
    list_result,
    axis=0,
    ignore_index=True
) #axis = 0 可以不写, ignore_index让index重新排序
result.tail(4)
```

#### 2.4.4 任务四：使用DataFrame自带的方法join方法和append：完成任务二和任务三的任务

```python
##  join 是 merging by index，但是需要column 不 overlap， 书里P253
result_up = df1.join(df3,how='inner') #默认左连接
result_down = df2.join(df4,how='inner') 

result = pd.concat(
    [result_up,result_down],
    axis=0,
    ignore_index=True
)

result.tail(4)

'''result = result_up.append(result_down,ignore_index=True)'''

```

自 pandas 1.4.0 版本后，append 方法已被弃用，推荐使用 concat 方法来替代。

#### 2.4.5 任务五：使用Panads的merge方法和DataFrame的append方法：完成任务二和任务三的任务

```python
#Merge
result_up = pd.merge(
    df1,df3,
    left_index=True,right_index=True
)
result_down = pd.merge(
    df2,df4,
    left_index=True,right_index=True
)

result = pd.concat(
    [result_up,result_down],
    axis=0,
    ignore_index=True
)

result.tail(4)
```

【思考】对比merge、join以及concat的方法的不同以及相同。思考一下在任务四和任务五的情况下，为什么都要求使用DataFrame的append方法，如何只要求使用merge或者join可不可以完成任务四和任务五呢？

#### 2.4.6 任务六：完成的数据保存为result.csv

```python
result.to_csv('result.csv')
```
